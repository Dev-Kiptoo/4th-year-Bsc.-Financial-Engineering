```python
import yfinance as yf
prices = yf.download(tickers="TSLA", start='2010-12-01', end="2023-12-01")[['Adj Close']].copy()
```

```python
print(prices)
```

```python
len(prices)
```

```python
# prices log returns
import numpy as np
prices['Log Rets'] = np.log(prices['Adj Close'] / prices['Adj Close'].shift(1))
prices
```

```python
# Daily and annual volatility
daily_vol = np.std(prices['Log Rets'])

# Annual volatility
Anual_vol =  daily_vol * 252 ** 0.5# 252 is the no of trading days
```

```python
daily_vol
Anual_vol
```

```python
import matplotlib.pyplot as plt
# plt.plot(prices['Adj Close'])
plt.plot(prices['Log Rets'])

plt.xlabel('Date')
plt.ylabel('Log returns')
plt.title('Open and Close Prices Over Time')
plt.legend()

plt.show()
```
